## Lab 6 - Bài 1

```python
import lab6_module as md

rectangles = [] 
squares = []  

while True:
    md.display_menu()
    choice = md.safe_num_input("Nhập vào lựa chọn của bạn: ")
    
    match choice:
        case 0:
            print("Tạm biệt!")
            break
        case 1: 
            rect = md.input_rectangle()
            rectangles.append(rect)
        case 2:
            sq = md.input_square()
            squares.append(sq)
        case 3: 
            md.print_rectangles(rectangles)
        case 4: 
            md.print_squares(squares)
        case 5:
            shapes = rectangles + squares
            md.print_shapes(shapes)
        case _:
            print("Chức nằng không tồn tại, vui lòng nhập lại!")
```

## Lab 6 - Bài 4

```python
from lab6_module import StudentManager

manager = StudentManager()  

while True:
    manager.display_menu()
    try:
        choice = int(input("Nhập lựa chọn của bạn: "))
    except ValueError:
        print("Lựa chọn không hợp lệ, vui lòng nhập số.")
        continue

    match choice:
        case 0:
            print("Tạm biệt!")
            break
        case 1:
            manager.add_student(manager.create_new_student())           
        case 2:
            manager.display_all_students()
        case 3:
            manager.display_good_students()
        case 4:
            manager.sort_students_by_gpa()
        case _:
            print("Chức năng không tồn tại, vui lòng nhập lại!")
```
