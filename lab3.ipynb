## Lab 3 - Bài 1
Viết chương trình thực hiện các **thao tác với một dãy số nguyên**.
- Sắp xếp theo thứ tự tăng dần và xuất dãy số nguyên ra màn hình.
- Xuất phần tử có giá trị nhỏ nhất ra màn hình.
- Tính và xuất ra màn hình trung bình cộng các phần tử chia hết cho 3.

```python


def safe_input_int(prompt: str) -> int:
    while True:
        try:
            return int(input(prompt))
        except ValueError:
            print("The input is invalid!, please reinput: ")

def input_list() -> list:
    my_list = []
    while True:
        list_range = safe_input_int("Input your list range: ")
        if list_range > 15:
            print("The list range is too big, please enter smaller number!")
        elif list_range <= 0:
            print("The list range must be greater than 0!")
        else:
            break     
    
    for i in range(list_range):
        num = safe_input_int(f"input your list item number {i+1}")
        my_list.append(num)  
    return my_list    

def quick_sort(arr:list) -> list:
    if len(arr) <= 1:
        return arr
    
    pivot = arr[-1]

    left = [x for x in arr[:-1] if x <= pivot]
    right = [x for x in arr[:-1] if x > pivot]

    return quick_sort(left) + [pivot] + quick_sort(right)

def avg_divide_by_three(arr) -> float:
    count_num = 0
    total = 0
    for i in arr:
        if i % 3 == 0:
            count_num += 1
            total += i
    if count_num == 0:
        return 0
    return total/count_num

sorted_list = quick_sort(input_list())
print(f"Sorted list is: {sorted_list}")
print(f"The smalest number in the list is: {sorted_list[0]}")
print(f"The average of numbers which divided evenly by three is: {avg_divide_by_three(sorted_list)}")
```

## Lab 3 - Bài 2
- Nhập vào số lượng sinh viên, nhập tên và điểm của từng sinh viên

```python
def safe_input_int(prompt: str) -> int:
    while True:
        try:
            return int(input(prompt))
        except ValueError:
            print("The input is invalid!, please reinput: ")

def safe_input_float(prompt: str) -> float:
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("The input is invalid!, please reinput: ")

def input_students() -> list:
    students = []

    number_student = safe_input_int("Enter number of students: ")

    for i in range(number_student):
        name = input(f"Enter student number {i+1} name: ")
        score = safe_input_float(f"Enter student number {i+1} score: ")

        students.append({"name": name, "score": score})
    return students

def student_classification(students: list) -> list:
    for student in students: 
        score = student.get("score", 0)  # lấy điểm, mặc định 0 nếu không có

        if score < 5:
            student["classification"] = "Yếu"
        elif 5 <= score < 7:
            student["classification"] = "Trung bình"
        elif 7 <= score < 8:
            student["classification"] = "Khá"
        elif 8 <= score < 9:
            student["classification"] = "Giỏi"
        else:  
            student["classification"] = "Xuất sắc"
    
    return students

# def find_student(students, search_name):
#     for student in students:
#         if student["name"] == search_name:
#             return student
#     return None

students = input_students()
print("Student data:", students)

# print("Finded student: ", find_student(students, students[1]["name"]))
# students.sort(key=lambda x: x["score"], reverse=True)  
print("student classification data: ", student_classification(students))
```

## Lab 3 - Bài 3

Nhập một số nguyên, kiểm tra có phải là số nguyên tố hay không?

```python
def safe_input_int(prompt: str) -> int:
    while True:
        try:
            return int(input(prompt))
        except ValueError:
            print("The input is invalid!, please reinput: ")

def is_prime(n: int) -> bool:
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

n = safe_input_int("Enter a integer: ")
if is_prime(n):
    print(f"{n} is a prime number!")
else: 
    print(f"{n} is not a prime number!")
```

## Lab 3 - Bài 5: bài tập làm thêm

```python
from datetime import datetime
import math, cmath

# menu
PHEP_TINH= (
    "Phép tính cơ bản",
    "Lũy thừa",
    "Căn bậc hai",
    "Hàm lượng giác",
    "Logarit",
    "Tìm số lớn nhất trong ba số",
    "Giải phương trình bậc 2",
    "Thời gian hiện tại",
    "Hiển thị lịch sử tính toán",
    "Thoát chương trình"
)

def safe_float_input(prompt: str) -> float:
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("Vui lòng nhập số hợp lệ.")

def nhap_so() -> tuple[float,float]:
    return safe_float_input("Nhập số thứ nhất (x): "), safe_float_input("Nhập số thứ hai (y): ")

def get_coefficients() -> tuple[float, float, float]:
    return safe_float_input("Nhập vào hệ số a: "), safe_float_input("Nhập vào hệ số b: "), safe_float_input("Nhập vào hệ số c: ")

def solve_linear(a: float, b: float) -> str:
    """Giải phương trình ax + b = 0"""
    if a == 0:
        if b == 0:
            return "Phương trình vô số nghiệm"
        else:
            return "Phương trình vô nghiệm"
    return f"Nghiệm của phương trình là: x = {-b/a}"

def calc_discriminant(a: float, b: float, c: float) -> tuple[float, float | complex]:
    """Tính delta và căn delta phương trình bậc 2"""
    delta = b**2 - 4*a*c
    sqrt_delta = math.sqrt(delta) if delta >= 0 else cmath.sqrt(delta)
    return delta, sqrt_delta

def solve_quadratic(a: float, b: float, c: float) -> str:
    """Giải phương trình bậc 2"""
    if a == 0:
        return solve_linear(b, c)
    
    delta, sqrt_delta = calc_discriminant(a, b, c)
    if delta == 0:
        x = -b / (2*a)
        return f"Phương trình có nghiệm kép: x = {x}"
    elif delta < 0:
        x1 = (-b + sqrt_delta) / (2*a)
        x2 = (-b - sqrt_delta) / (2*a)
        return f"Phương trình có 2 nghiệm phức phân biệt:\nx1 = {x1}\nx2 = {x2}"
    else:
        x1 = (-b + sqrt_delta) / (2*a)
        x2 = (-b - sqrt_delta) / (2*a)
        return f"Phương trình có 2 nghiệm thực phân biệt:\nx1 = {x1}\nx2 = {x2}"

def input_history(history:list, cal: str, input: str, output: str) -> list:
    history.append({"calculation": cal, "input": input, "output": output})
    return history

def print_history(history: list) -> None:
    print("\n--- Calculation History ---")
    if not history:
        print("Chưa có lịch sử tính toán nào.")
    else:
        # in tiêu đề cột
        print(f"{'STT':<5} {'Phép tính':<30} {'Input':<30} {'Output':<30}")
        print("-" * 120)
        # in từng dòng lịch sử
        for i, record in enumerate(history, start=1):
            print(f"{i:<5} {record['calculation']:<30} {record['input']:<30} {record['output']:<30}")

def chon_phep_tinh() -> int:
    print("\n--- Simple calculator ---")
    for i, pt in enumerate(PHEP_TINH, start=1):
        print(f"{i}. {pt}")
    return int(input("Chọn phép tính: "))

history = []

while True:
    choose = chon_phep_tinh()
    match choose:
        case 1:
            print("\n--- 1. Phép tính cơ bản ---")
            x, y = nhap_so()
            out1 = f"{x} + {y} = {x + y}"
            out2 = f"{x} - {y} = {x - y}"
            out3 = f"{x} * {y} = {x * y}"
            out4 = f"{x} / {y} = {x / y}" if y != 0 else "Không thể chia cho 0"
            print(out1, out2, out3, out4, sep="\n")
            input_history(history, "Phép tính cơ bản", f"{x}, {y}", f"{out1}, {out2}, {out3}, {out4}")

        case 2:
            x, y = nhap_so()
            out = f"{x}^{y} = {x ** y}"
            print(out)
            input_history(history, "Lũy thừa", f"{x}, {y}", out)

        case 3:
            x = safe_float_input("Nhập số cần tính: ")
            out = f"√{x} = {math.sqrt(x) if x >= 0 else cmath.sqrt(x)}"
            print(out)
            input_history(history, "Căn bậc hai", str(x), out)

        case 4:
            x = safe_float_input("Nhập số cần tính (radian): ")
            out1 = f"sin({x}) = {math.sin(x)}"
            out2 = f"cos({x}) = {math.cos(x)}"
            out3 = f"tan({x}) = {math.tan(x)}"
            print(out1, out2, out3, sep="\n")
            input_history(history, "Hàm lượng giác", str(x), f"{out1}, {out2}, {out3}")

        case 5:
            sub_log = int(input("1. Log10 & ln | 2. Log cơ số tùy chọn: "))
            if sub_log == 1:
                x = safe_float_input("Nhập số cần tính: ")
                out1 = f"log10({x}) = {math.log10(x)}" if x > 0 else "Không thể tính logarit của số không dương"
                out2 = f"ln({x}) = {math.log(x)}" if x > 0 else ""
                print(out1, out2, sep="\n")
                input_history(history, "Logarit (log10, ln)", str(x), f"{out1}, {out2}")
            elif sub_log == 2:
                x, y = nhap_so()
                out = f"log cơ số {y} ({x}) = {math.log(x, y)}" if x > 0 and y > 0 and y != 1 else "Cơ số log không hợp lệ"
                print(out)
                input_history(history, "Logarit cơ số tùy chọn", f"{x}, {y}", out)

        case 6:
            print("--- Tìm số lớn nhất trong 3 số ---")
            a = float(input("Nhập số thứ nhất: "))
            b = float(input("Nhập số thứ hai: "))
            c = float(input("Nhập số thứ ba: "))

            max_value = a 
            if b > max_value:
                max_value = b
            if c > max_value:
                max_value = c
            out = f"Số lớn nhất là số: {max_value}"
            print(out)
            input_history(history, "Tìm số lớn nhất trong ba số", f"{a}, {b}, {c}", out)

        case 7:
            print("--- Giải phương trình bậc 2 ---")
            a, b, c = get_coefficients()
            out = solve_quadratic(a, b, c)
            print(out)
            input_history(history, "Giải phương trình bậc 2", f"{a}, {b}, {c}", str(out))

        case 8:
            now = datetime.now()
            out = f"Thời gian hiện tại là: {now}"
            print(out)
            input_history(history, "Thời gian hiện tại", "", out)

        case 9:
            print_history(history)

        case 10:
            print("Chào tạm biệt!")
            break
        case _:
            print("Lựa chọn không hợp lệ.")

```
