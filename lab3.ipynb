## Lab 3 - Bài 1
Viết chương trình thực hiện các **thao tác với một dãy số nguyên**.
- Sắp xếp theo thứ tự tăng dần và xuất dãy số nguyên ra màn hình.
- Xuất phần tử có giá trị nhỏ nhất ra màn hình.
- Tính và xuất ra màn hình trung bình cộng các phần tử chia hết cho 3.

```python

my_list = []
# khai báo số lượng phần tử có trong list
while True:
    list_range = int(input("Input your list range: "))
    if list_range > 15:
        print("The list range is too big, please enter smaller number!")
    elif list_range <= 0:
        print("The list range must be greater than 0!")
    else:
        break     
# Input số vào list từ bàn phím
for i in range(list_range):
    num = float(input(f"input your list item number {i+1}"))
    my_list.append(num)  

# tính trung bình cộng các số chia hết cho 3
count_num = 0
total = 0
for i in my_list:
    if i % 3 == 0:
        count_num += 1
        total += i
if count_num == 0:
    print("Không có số nào chia hết cho 3.")
else:
    print(f"Trung bình cộng các số chia hết cho 3 bằng: {total/count_num}")

# xắp xếp lại list và xuất la màn hình 
my_list.sort(reverse=False)
print(f"Chuỗi sau xắp xếp: {my_list}")
print(f"Số bé nhất trong chuỗi là: {my_list[0]}")


```

## Lab 3 - Bài 2
- Nhập vào số lượng sinh viên, nhập tên và điểm của từng sinh viên
- Xếp loại sinh viên dựa trên điểm đã nhập
- Xuất ra màn hình danh sách sinh viên đã xếp loại

```python

students = []

while True:
    print("\n--- MENU ---")
    print("0. Thoát chương trình")
    print("1. Nhập thông tin sinh viên")
    print("2. Xuất thông tin sinh viên")

    # Nhập lựa chọn (kiểm tra lỗi)
    while True:
        try:
            choose = int(input("Nhập lựa chọn của bạn: "))
            break
        except ValueError:
            print("The input is invalid!, please reinput: ")

    # Thoát chương trình
    if choose == 0:
        print("Thank you!")
        break

    # Nhập thông tin sinh viên
    elif choose == 1:
        while True:
            try:
                number_student = int(input("Enter number of students: "))
                break
            except ValueError:
                print("The input is invalid!, please reinput: ")

        students = []

        for i in range(number_student):
            name = input(f"Enter student number {i+1} name: ")

            while True:
                try:
                    score = float(input(f"Enter student number {i+1} score: "))
                    break
                except ValueError:
                    print("The input is invalid!, please reinput: ")

            # Phân loại sinh viên theo điểm
            if score < 5:
                classification = "Yếu"
            elif 5 <= score < 7:
                classification = "Trung bình"
            elif 7 <= score < 8:
                classification = "Khá"
            elif 8 <= score < 9:
                classification = "Giỏi"
            else:
                classification = "Xuất sắc"

            students.append({
                "name": name,
                "score": score,
                "classification": classification
            })

        # Sắp xếp theo điểm giảm dần
        students.sort(key=lambda x: x["score"], reverse=True)
        print("Nhập thông tin sinh viên thành công!")

    # Xuất thông tin sinh viên
    elif choose == 2:
        if not students:
            print("Chưa có dữ liệu sinh viên! Vui lòng nhập trước.")
        else:
            print("\n--- Danh sách sinh viên ---")
            for student in students:
                print(f"Tên: {student['name']}, "
                      f"Điểm: {student['score']}, "
                      f"Xếp loại: {student['classification']}")
    # Trường hợp nhập sai lựa chọn
    else:
        print("Lựa chọn không hợp lệ, vui lòng chọn lại.")
```

## Lab 3 - Bài 3

Nhập một số nguyên, kiểm tra có phải là số nguyên tố hay không?

```python

# Nhập số nguyên, có kiểm tra lỗi
while True:
    try:
        n = int(input("Enter an integer: "))
        break
    except ValueError:
        print("The input is invalid!, please reinput: ")

# Kiểm tra số nguyên tố
if n < 2:
    print(f"{n} is not a prime number!")
else:
    is_prime = True
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            is_prime = False
            break

    if is_prime:
        print(f"{n} is a prime number!")
    else:
        print(f"{n} is not a prime number!")
```

Phần bên dưới là phần làm thêm cho lịch sử, dự định sẽ lưu lịch sử trong một file csv để khi kết thức chương trình lịch sử vẫn được lưu lại.

```python
from datetime import datetime
import math, cmath

# Danh sách các phép tính
PHEP_TINH = (
    "Phép tính cơ bản",
    "Lũy thừa",
    "Căn bậc hai",
    "Hàm lượng giác",
    "Logarit",
    "Tìm số lớn nhất trong ba số",
    "Giải phương trình bậc 2",
    "Thời gian hiện tại",
    "Hiển thị lịch sử tính toán",
    "Thoát chương trình"
)

history = []

while True:
    print("\n--- Simple Calculator ---")
    for i, pt in enumerate(PHEP_TINH, start=1):
        print(f"{i}. {pt}")

    # Nhập lựa chọn
    while True:
        try:
            choose = int(input("Chọn phép tính: "))
            break
        except ValueError:
            print("Vui lòng nhập số hợp lệ!")

    # 1. Phép tính cơ bản
    if choose == 1:
        print("\n--- 1. Phép tính cơ bản ---")
        while True:
            try:
                x = float(input("Nhập số thứ nhất (x): "))
                y = float(input("Nhập số thứ hai (y): "))
                break
            except ValueError:
                print("Vui lòng nhập số hợp lệ!")

        out1 = f"{x} + {y} = {x + y}"
        out2 = f"{x} - {y} = {x - y}"
        out3 = f"{x} * {y} = {x * y}"
        out4 = f"{x} / {y} = {x / y}" if y != 0 else "Không thể chia cho 0"
        print(out1, out2, out3, out4, sep="\n")
        history.append({"calculation": "Phép tính cơ bản", "input": f"{x}, {y}", "output": f"{out1}, {out2}, {out3}, {out4}"})

    # 2. Lũy thừa
    elif choose == 2:
        while True:
            try:
                x = float(input("Nhập cơ số (x): "))
                y = float(input("Nhập số mũ (y): "))
                break
            except ValueError:
                print("Vui lòng nhập số hợp lệ!")
        out = f"{x}^{y} = {x ** y}"
        print(out)
        history.append({"calculation": "Lũy thừa", "input": f"{x}, {y}", "output": out})

    # 3. Căn bậc hai
    elif choose == 3:
        while True:
            try:
                x = float(input("Nhập số cần tính: "))
                break
            except ValueError:
                print("Vui lòng nhập số hợp lệ!")
        out = f"√{x} = {math.sqrt(x) if x >= 0 else cmath.sqrt(x)}"
        print(out)
        history.append({"calculation": "Căn bậc hai", "input": str(x), "output": out})

    # 4. Hàm lượng giác
    elif choose == 4:
        while True:
            try:
                x = float(input("Nhập số cần tính (radian): "))
                break
            except ValueError:
                print("Vui lòng nhập số hợp lệ!")
        out1 = f"sin({x}) = {math.sin(x)}"
        out2 = f"cos({x}) = {math.cos(x)}"
        out3 = f"tan({x}) = {math.tan(x)}"
        print(out1, out2, out3, sep="\n")
        history.append({"calculation": "Hàm lượng giác", "input": str(x), "output": f"{out1}, {out2}, {out3}"})

    # 5. Logarit
    elif choose == 5:
        while True:
            try:
                sub_log = int(input("1. Log10 & ln | 2. Log cơ số tùy chọn: "))
                break
            except ValueError:
                print("Vui lòng nhập số hợp lệ!")

        if sub_log == 1:
            while True:
                try:
                    x = float(input("Nhập số cần tính: "))
                    break
                except ValueError:
                    print("Vui lòng nhập số hợp lệ!")
            if x > 0:
                out1 = f"log10({x}) = {math.log10(x)}"
                out2 = f"ln({x}) = {math.log(x)}"
            else:
                out1 = "Không thể tính logarit của số không dương"
                out2 = ""
            print(out1, out2, sep="\n")
            history.append({"calculation": "Logarit (log10, ln)", "input": str(x), "output": f"{out1}, {out2}"})
        elif sub_log == 2:
            while True:
                try:
                    x = float(input("Nhập số cần tính: "))
                    y = float(input("Nhập cơ số: "))
                    break
                except ValueError:
                    print("Vui lòng nhập số hợp lệ!")
            if x > 0 and y > 0 and y != 1:
                out = f"log cơ số {y} ({x}) = {math.log(x, y)}"
            else:
                out = "Cơ số log không hợp lệ"
            print(out)
            history.append({"calculation": "Logarit cơ số tùy chọn", "input": f"{x}, {y}", "output": out})

    # 6. Tìm số lớn nhất trong ba số
    elif choose == 6:
        print("--- Tìm số lớn nhất trong 3 số ---")
        while True:
            try:
                a = float(input("Nhập số thứ nhất: "))
                b = float(input("Nhập số thứ hai: "))
                c = float(input("Nhập số thứ ba: "))
                break
            except ValueError:
                print("Vui lòng nhập số hợp lệ!")
        max_value = a
        if b > max_value:
            max_value = b
        if c > max_value:
            max_value = c
        out = f"Số lớn nhất là: {max_value}"
        print(out)
        history.append({"calculation": "Tìm số lớn nhất trong ba số", "input": f"{a}, {b}, {c}", "output": out})

    # 7. Giải phương trình bậc 2
    elif choose == 7:
        print("--- Giải phương trình bậc 2 ---")
        while True:
            try:
                a = float(input("Nhập hệ số a: "))
                b = float(input("Nhập hệ số b: "))
                c = float(input("Nhập hệ số c: "))
                break
            except ValueError:
                print("Vui lòng nhập số hợp lệ!")

        if a == 0:
            if b == 0:
                out = "Phương trình vô nghiệm" if c != 0 else "Phương trình vô số nghiệm"
            else:
                out = f"Nghiệm của phương trình là: x = {-c / b}"
        else:
            delta = b**2 - 4*a*c
            sqrt_delta = math.sqrt(delta) if delta >= 0 else cmath.sqrt(delta)
            if delta == 0:
                x = -b / (2*a)
                out = f"Phương trình có nghiệm kép: x = {x}"
            elif delta < 0:
                x1 = (-b + sqrt_delta) / (2*a)
                x2 = (-b - sqrt_delta) / (2*a)
                out = f"Phương trình có 2 nghiệm phức phân biệt:\nx1 = {x1}\nx2 = {x2}"
            else:
                x1 = (-b + sqrt_delta) / (2*a)
                x2 = (-b - sqrt_delta) / (2*a)
                out = f"Phương trình có 2 nghiệm thực phân biệt:\nx1 = {x1}\nx2 = {x2}"
        print(out)
        history.append({"calculation": "Giải phương trình bậc 2", "input": f"{a}, {b}, {c}", "output": out})

    # 8. Thời gian hiện tại
    elif choose == 8:
        now = datetime.now()
        out = f"Thời gian hiện tại là: {now}"
        print(out)
        history.append({"calculation": "Thời gian hiện tại", "input": "", "output": out})

    # 9. Hiển thị lịch sử tính toán
    elif choose == 9:
        print("\n--- Calculation History ---")
        if not history:
            print("Chưa có lịch sử tính toán nào.")
        else:
            print(f"{'STT':<5} {'Phép tính':<30} {'Input':<30} {'Output':<30}")
            print("-" * 120)
            for i, record in enumerate(history, start=1):
                print(f"{i:<5} {record['calculation']:<30} {record['input']:<30} {record['output']:<30}")

    # 10. Thoát chương trình
    elif choose == 10:
        print("Chào tạm biệt!")
        break

    else:
        print("Lựa chọn không hợp lệ!")
```
